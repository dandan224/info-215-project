```python
df = listings[['neighbourhood','number_of_reviews','review_scores_rating','review_scores_value','review_scores_accuracy','review_scores_cleanliness',\
             'review_scores_communication','review_scores_checkin','review_scores_location','calculated_host_listings_count']]
```

```python
df = df[df['number_of_reviews']>0]
```

```python
df['number_of_reviews'].value_counts()
```

```python
df.isnull().sum()
```

```python
df[['review_scores_rating','review_scores_value','review_scores_accuracy','review_scores_cleanliness',\
   'review_scores_communication','review_scores_checkin','review_scores_location']]\
= df[['review_scores_rating','review_scores_value','review_scores_accuracy','review_scores_cleanliness',\
   'review_scores_communication','review_scores_checkin','review_scores_location']].\
fillna(round(df[['review_scores_rating','review_scores_value','review_scores_accuracy','review_scores_cleanliness',
   'review_scores_communication','review_scores_checkin','review_scores_location']].mean()))
```

```python
df.isnull().sum()
# the null of neighbourhood does not impact the result, so here do not need to clean these null values.
```

```python
# try to do the feature-importance without flatten

from sklearn.model_selection import cross_val_score, StratifiedKFold
from sklearn.tree import DecisionTreeClassifier
import matplotlib.pyplot as plt
X = df.drop(['review_scores_rating','number_of_reviews','neighbourhood'], axis = 1)
y = df['review_scores_rating']
model = DecisionTreeClassifier()
cv = StratifiedKFold(n_splits=5, shuffle=True)
s = cross_val_score(model, X, y, cv=cv, n_jobs=-1)
print(s)
print("Score:\t{:0.3} ± {:0.3}".format(s.mean().round(3), s.std().round(3)))

model.fit(X,y)
print(model.feature_importances_) #use inbuilt class feature_importances of tree based classifiers
#plot graph of feature importances for better visualization
feat_importances = pd.Series(model.feature_importances_, index=X.columns)
feat_importances.nlargest(10).plot(kind='barh', title = 'The importance of features to the rating')
plt.show()
```

```python
# try to do the feature-importance with flatten
df_f = df.loc[df.index.repeat(df.number_of_reviews)]
X = df_f.drop(['review_scores_rating','number_of_reviews','neighbourhood'], axis = 1)
y = df_f['review_scores_rating']
model = DecisionTreeClassifier()
cv = StratifiedKFold(n_splits=5, shuffle=True)
s = cross_val_score(model, X, y, cv=cv, n_jobs=-1)
print(s)
print("Score:\t{:0.3} ± {:0.3}".format(s.mean().round(3), s.std().round(3)))

model.fit(X,y)
print(model.feature_importances_) #use inbuilt class feature_importances of tree based classifiers
#plot graph of feature importances for better visualization
feat_importances = pd.Series(model.feature_importances_, index=X.columns)
feat_importances.nlargest(10).plot(kind='barh', title = 'The importance of features to the rating for Paris')
plt.show()
```

```python
import string
reviews_list = [ str(x).strip().lower() for x in list(reviews_new.comments.values) ]
clean_reviews_list = []
punct = set(string.punctuation)
for review in reviews_list:
    clean_reviews_list.append(''.join([c for c in review if c not in punct]))
```

```python
wordcloud = WordCloud(max_font_size=100, max_words=300, background_color="white", width=1000, height=600).generate(' '.join(clean_reviews_list))

# Display the generated image:
plt.figure(figsize=[20,10])
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()
```

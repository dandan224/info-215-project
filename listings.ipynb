```python
#First, remove all columns with more than 50% missing values.
#Next, for the attributes with more than 30% missing values, we will impute them individually.
#Some attributes like the reviews related columns have almost 20% missing values, we will impute them with the mean value. 
# And if within 5%, We will handle these by replacing these values with their corresponding mode
miss_listings = listings.isnull().sum() / len(listings)
miss_gr_05 = listings.columns[miss_listings > 0.5]
listings = listings.drop(miss_gr_05, axis=1)
miss_listings
```

```python
# host_about,transit features, neighborhood_overview, and space are are freeform text values. 
# Since I will not be doing any natural language processing, we can drop these.
listings = listings.drop(['host_about','transit', 'neighborhood_overview','space'], axis=1)
```

```python
# Currency attributes can be converted to Floats by removing the dollar symbol and commas 
listings[['extra_people','price','cleaning_fee', 'security_deposit']] = \
listings[[ 'extra_people','price','cleaning_fee','security_deposit']].\
           apply(lambda col: col.str.replace("$", "").str.replace(",","").astype('float'))
```

```python
listings[['cleaning_fee', 'security_deposit']] = listings[['cleaning_fee', 'security_deposit']].\
                     fillna(listings[['cleaning_fee', 'security_deposit']].mean())
```

```python
from sklearn.model_selection import cross_val_score, StratifiedKFold
from sklearn.tree import DecisionTreeClassifier
import matplotlib.pyplot as plt
X = df.drop('review_scores_rating', axis = 1)
y = df['review_scores_rating']
model = DecisionTreeClassifier()
cv = StratifiedKFold(n_splits=5, shuffle=True)
s = cross_val_score(model, X, y, cv=cv, n_jobs=-1)
print(s)
print("Score:\t{:0.3} ± {:0.3}".format(s.mean().round(3), s.std().round(3)))

model.fit(X,y)
print(model.feature_importances_) #use inbuilt class feature_importances of tree based classifiers
#plot graph of feature importances for better visualization
feat_importances = pd.Series(model.feature_importances_, index=X.columns)
feat_importances.nlargest(10).plot(kind='barh', title = 'The importance of features to the rating')
plt.show()
```

```python
from wordcloud import WordCloud, STOPWORDS
from PIL import Image
import numpy as np
import urllib
import requests
import matplotlib.pyplot as plt
```

```python
import string
reviews_list = [ str(x).strip().lower() for x in list(reviews_rate.comments.values) ]
clean_reviews_list = []
punct = set(string.punctuation)
for review in reviews_list:
    clean_reviews_list.append(''.join([c for c in review if c not in punct]))
```

```python
wordcloud = WordCloud(max_font_size=100, max_words=300, background_color="white", width=1000, height=600).generate(' '.join(clean_reviews_list))

# Display the generated image:
plt.figure(figsize=[20,10])
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()
```
